```python
# convert mp4 to images
# from Stack Overflow
import cv2
vidcap = cv2.VideoCapture('project_output4.mp4')
success,image = vidcap.read()
count = 0
while success:
  cv2.imwrite("captures/pframe%d.jpg" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
